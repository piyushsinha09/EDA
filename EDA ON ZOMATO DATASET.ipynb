# Data visualisation

#plotting pie chart

```python
country_name= df_new.Country.value_counts().index
```

```python
country_vals= df_new.Country.value_counts().values
```

```python
plt.pie(country_vals[:3], labels= country_name[:3], autopct= '%1.2f%%') #getting for top 3 countries where zomato transaction is highest
```

```python
df_new.columns
```

```python
df_new.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings= df_new.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns= {0:'Rating counts'})
ratings
```

It shows white: no rating, red:poor, orange:average, yellow: good, green:very good, dark green: excellent

```python
#constructing barplot between agrregate rating and rating counts
plt.rcParams['figure.figsize']=(8,6)
sns.barplot(x='Aggregate rating', y='Rating counts',data= ratings)
```

```python
plt.rcParams['figure.figsize']=(8,6)
sns.barplot(x='Aggregate rating', y='Rating counts',hue='Rating color',data= ratings,palette=['blue','red','orange','yellow','green','green'])
```

OBSERVATIONS:
    customer rating 0:white------>not rated 
    customer rating 1.8-2.4: red------>poor
    customer rating 2.5-3.4: orange---->average
    customer rating 3.5- 3.9: yellow----->good
    customer rating 4.0- 4.4: green------->very good
    customer rating 4.5- 5.0: dark green---->excellent

```python
#count plot
sns.countplot(x='Rating color',data= ratings,palette=['blue','red','orange','yellow','green','green'])
```
